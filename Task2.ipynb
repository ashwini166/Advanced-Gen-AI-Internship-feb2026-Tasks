Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

```python
# Take input from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.

```python
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Print result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values Problem Statement Find the maximum and minimum values from a list of numbers.

```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage Problem Statement Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count number of Present days
present_days = attendance.count("P")

# Calculate percentage
percentage = (present_days / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```
